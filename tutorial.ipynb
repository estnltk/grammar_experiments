# Finite grammar experiments

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_graph(graph):
    pos = nx.drawing.nx_pydot.pydot_layout(graph, prog='dot')
    plt.figure(figsize=(8,8)) 
    nx.draw(graph, with_labels=True, pos=pos)
    plt.show()
```

```python
from trees import  graph_from_document
document = {'F':[(1,4),(5,10),          (17,19)], 
            'G':[(1,4),       (11, 13), (15,19)]}

graph = graph_from_document(document)
plot_graph(graph)
```

```python
from trees import Rule, Grammar
rules = []
rules.append(Rule('S', 'A'))
rules.append(Rule('S', 'B'))
rules.append(Rule('A', 'BF'))
rules.append(Rule('B', 'G'))

grammar = Grammar(start_symbol='S', rules=rules)
grammar
```

```python
from trees import document_to_graph
result = document_to_graph(document, grammar)
plot_graph(result)
```
